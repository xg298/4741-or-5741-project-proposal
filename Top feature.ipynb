```python
data_model1 = {'Model': ['F-test']*10, 'Top_Feature': ['Median Gross Rent','housewithmortgage','infoserviceperc','detachedhouse','wkathm','TotalPopulationForeignBorn','FoodEnvironmentIndex','PrimaryCarePhysiciansPCPRa','notcitizen','housecost30inc'], 'Importance': [3.4031,2.2696,1.2053,0.8371,0.8091,0.6887,0.5268,0.5065,0.3717,0.3613]}
data_model2 = {'Model': ['SGDRegressor']*10, 'Top_Feature': ['Median Gross Rent','housewithmortgage','infoserviceperc','wkathm','detachedhouse','TotalPopulationForeignBorn','PrimaryCarePhysiciansPCPRa','FoodEnvironmentIndex','notcitizen','housecost30inc'], 'Importance': [0.6895,0.6011,0.4044,0.3629,0.3409,0.3267,0.2874,0.2009,0.1986,0.1930]}
data_model3 = {'Model': ['Treemodel']*10, 'Top_Feature': ['Median Gross Rent','housewithmortgage','infoserviceperc','wkathm','detachedhouse','TotalPopulationForeignBorn','PopulationDensityPerSqMile','PrimaryCarePhysiciansPCPRa','FoodEnvironmentIndex', 'PeopleinSubsidizedHousing'], 'Importance': [0.2801, 0.2118, 0.0945, 0.0555, 0.0540, 0.0455, 0.0303, 0.0249, 0.0231, 0.0230]}
```

```python
df_model1 = pd.DataFrame(data_model1)
df_model2 = pd.DataFrame(data_model2)
df_model3 = pd.DataFrame(data_model3)

# 合并数据框
merged_df = pd.concat([df_model1, df_model2, df_model3])

plt.figure(figsize=(12, 8))
sns.scatterplot(x='Top_Feature', y='Importance', hue='Model', data=merged_df, palette='viridis',size=100)
plt.title('Top Features in Different Models')
plt.xlabel('Top Feature')
plt.ylabel('Importance')
plt.xticks(rotation=45, ha='right')  # 旋转 x 轴标签，使其更易读
plt.show()
```
